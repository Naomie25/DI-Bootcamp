 Exercise 2 : Transformer Architecture and Tokenization

Tokenization is the process of converting text into smaller units (called tokens) that a machine learning model can understand.

```python
from transformers import AutoTokenizer
import matplotlib.pyplot as plt

# Texte
text = "Hi my name is Naomie, I am 25 years old"

# 2. Tokenize input text
# Utiliser le tokenizer de BERT
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

# Tokenisation
tokens = tokenizer.tokenize(text)
token_ids = tokenizer.convert_tokens_to_ids(tokens)
```

```python
# 3. Visualizing the tokenization process
plt.figure(figsize=(10, 4))
plt.bar(tokens, token_ids, color="skyblue")
plt.xlabel("Tokens ")
plt.ylabel("Token IDs ")
plt.title("Tokenization Visualization ")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Exercise 3 : Understanding Token IDs and Special Prefixes

```python
#Print tokens with their IDs
for token, token_id in zip(tokens, token_ids):
    print(f"Token: '{token}' \t ID: {token_id}")
```

Prefixes like 'Ġ' indicate that the token starts after a space, helping the model preserve the structure of natural text.

Exercise 4 : Pretraining vs. Fine-Tuning

Pretraining:
Train a language model on a large general text dataset to learn language patterns and understanding without any specific task in mind.

Fine-Tuning:
Take the pretrained model and train it further on a smaller, task-specific dataset to make it perform well on that particular task (like sentiment analysis or summarization).

Exercise 5 : Generate Simple Text

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

# Initialize model and tokenizer
model_name = "gpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Create a variable with your input sentence
input_text = "Artificial intelligence is transforming"

# Encode input text and generate output token IDs
input_ids = tokenizer.encode(input_text, return_tensors="pt")
output_ids = model.generate(input_ids, max_length=50, num_return_sequences=1, no_repeat_ngram_size=2)

# Decode the generated token IDs back to text
output_text = tokenizer.decode(output_ids[0], skip_special_tokens=True)

print(f"Input: {input_text}")
print(f"Generated Output: {output_text}")
```
