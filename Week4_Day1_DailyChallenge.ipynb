```python
import numpy as np
data=np.random.uniform(-5, 35, size=(10,12))
data
```

```python
import pandas as pd
cities = [
    "Paris", "Madrid", "Tel Aviv", "Lisbonne", "Milan",
    "CasaBlanca", "Marrakesh", "Cancun", "Montreal", "Marseille"
]

# Column labels (months)
months = [
    "January", "February", "March", "April",
    "May", "June", "July", "August",
    "September", "October", "November", "December"
]

# Create the DataFrame
df = pd.DataFrame(data, index=cities, columns=months)
df
```

```python
#Calculate the annual average temperature
annual_avg_temp = df.mean(axis=1)
print(annual_avg_temp)
# City with highest average temperature
city_highest = annual_avg_temp.idxmax()

# City with lowest average temperature
city_lowest = annual_avg_temp.idxmin()

print(f"City with highest average temperature: {city_highest}")
print(f"City with lowest average temperature: {city_lowest}")

#Calculate the annual average temperature for each city
annual_avg_temp_each_city = df.mean(axis=0)
print(annual_avg_temp_each_city)

# Highest and lowest average temperature for the year.
temp_highest = annual_avg_temp_each_city .idxmax()
temp_lowest = annual_avg_temp_each_city .idxmin()

print(f"City with highest average temperature: {temp_highest}")
print(f"City with lowest average temperature: {temp_lowest}")
```

```python
# Data visualization
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 7))

# Plot each city line
for city in df.index:
    plt.plot(df.columns, df.loc[city], marker='o', label=city)

plt.title('Monthly Temperature Trends by City')
plt.xlabel('Month')
plt.ylabel('Temperature (Â°C)')
plt.xticks(rotation=45)  # Rotate month labels for readability
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))
plt.grid(True)
plt.tight_layout()
plt.show()
```
