Exercise 1: Understanding Data Visualization

- Data visualization helps simplify complex data, making it easier to understand and communicate insights.

- It reveals patterns, trends, and outliers that may not be visible in raw numbers.

- Visualization supports better decision-making by presenting data in a clear and accessible way.

- A line graph shows how data changes over time.

- It is used to track trends, fluctuations, and comparisons across continuous variables.

- Ideal for time series data (e.g., sales per month, temperature per day).

Exercise 2: Creating a Line Plot for Temperature Variation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
temperatures = [72, 74, 76, 80, 82, 78, 75]

plt.plot(days, temperatures, marker='o', linestyle='-', color='blue')

plt.xlabel('Day')
plt.ylabel('Temperature (Â°F)')
plt.title('Temperature Variations Over a Week')

# Show the plot
plt.grid(True)
plt.show()
```

Exercise 3: Visualizing Monthly Sales with a Bar Chart

```python
months = [
    "January", "February", "March", "April", "May", "June",
    "July", "August", "September", "October", "November", "December"
]

sales = [5000, 5500, 6200, 7000, 7500, 6800,
         7200, 7600, 8100, 8300, 7900, 7700]  # 12 values

plt.bar(months, sales, color='pink')
plt.xlabel('Month')
plt.ylabel('Sales Amount ($)')
plt.title('Sales by Month (Full Year)')
plt.xticks(rotation=45)           # optional: tilt labels for readability
plt.tight_layout()                # optional: adjust layout
plt.show()
```

Exercise 5: Comparing Anxiety Levels Across Different Genders

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Do you have Anxiety?': ['Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No',
                             'Yes', 'No', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes'],
    'Choose your gender': ['Male', 'Female', 'Female', 'Male', 'Male', 'Female', 'Male', 'Other', 'Female', 'Male',
                          'Female', 'Male', 'Other', 'Female', 'Female', 'Male', 'Other', 'Female', 'Male', 'Other']
}

df = pd.DataFrame(data)

# Calculer la proportion d'anxiÃ©tÃ© par genre
proportions = (
    df.groupby(['Choose your gender', 'Do you have Anxiety?'])
      .size()
      .reset_index(name='count')
)

total_per_gender = proportions.groupby('Choose your gender')['count'].transform('sum')
proportions['proportion'] = proportions['count'] / total_per_gender

# Filtrer pour ne garder que les "Yes" (anxiÃ©tÃ© prÃ©sente)
anxiety_proportion = proportions[proportions['Do you have Anxiety?'] == 'Yes']

# Tracer le bar plot
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.barplot(data=anxiety_proportion, x='Choose your gender', y='proportion', palette='coolwarm')

plt.title("Proportion d'Ã©tudiants avec anxiÃ©tÃ© selon le genre", fontsize=16)
plt.ylabel("Proportion")
plt.xlabel("Genre")
plt.ylim(0, 1)

plt.show()

```

Exercise 6: Exploring the Relationship Between Age and Panic Attacks

```python
np.random.seed(42)
ages = np.random.randint(18, 61, size=100)

# GÃ©nÃ©rer alÃ©atoirement des rÃ©ponses Oui/Non pour les attaques de panique
panic_attacks = np.random.choice(['Yes', 'No'], size=100, p=[0.3, 0.7])  # 30% Oui, 70% Non

# CrÃ©er le DataFrame
df = pd.DataFrame({
    'Age': ages,
    'Do you have Panic Attacks?': panic_attacks
})
df['Panic_Attacks_Num'] = df['Do you have Panic Attacks?'].map({'Yes': 1, 'No': 0})


sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

sns.scatterplot(
    data=df,
    x='Age',
    y='Panic_Attacks_Num',
    hue='Do you have Panic Attacks?',   # Pour colorer selon la rÃ©ponse
    palette={'Yes': 'red', 'No': 'blue'},
    alpha=0.7,
    s=60,  # taille des points
    edgecolor='k'
)

# --- Personnalisation du graphique ---
plt.title("Relation entre l'Ã¢ge et la prÃ©sence d'attaques de panique", fontsize=16)
plt.xlabel("Ãge")
plt.ylabel("PrÃ©sence d'attaques de panique (0 = Non, 1 = Oui)")
plt.yticks([0, 1], ['Non', 'Oui'])  # Remplacer les ticks y par Oui/Non

plt.legend(title='Attaques de panique')
plt.tight_layout()
plt.show()
```
