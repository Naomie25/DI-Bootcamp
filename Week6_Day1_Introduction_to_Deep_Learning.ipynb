```python
# Input features
x1 = 50  # Study hours
x2 = 70  # Previous test score

# True label
y_true = 85  # Actual final exam score

# Model parameters (initial)
w1 = 0.6
w2 = 0.3
b = 10

# Learning rate
alpha = 0.0001
```

image.png

```python
# Step 1: Forward Pass - Make prediction
y_pred = w1 * x1 + w2 * x2 + b
print("Predicted Score ():", y_pred)

# Step 2: Compute error and loss (MSE with factor 1/2)
error = y_pred - y_true
loss = 0.5 * (error ** 2)
print("Error:", error)
print("Loss:", loss)
```

image.png

```python
# Step 3: Compute Gradients (partial derivatives)
dw1 = error * x1
dw2 = error * x2
db = error

print("Gradient w.r.t w1:", dw1)
print("Gradient w.r.t w2:", dw2)
print("Gradient w.r.t b:", db)
```

image.png

```python
# Step 4: Update weights and bias using gradient descent
w1 = w1 - alpha * dw1
w2 = w2 - alpha * dw2
b = b - alpha * db
y_pred = w1 * x1 + w2 * x2 + b
# Print updated parameters
print("\nUpdated Parameters:")
print("w1 =", w1)
print("w2 =", w2)
print("b  =", b)
print(y_pred)
```

```python
# Original input values
x1_raw = 50   # Study hours
x2_raw = 70   # Previous test score

# Min-Max Normalization (scale between 0 and 1)
# You can replace min/max values with actual dataset min/max if available
x1_min, x1_max = 0, 100
x2_min, x2_max = 0, 100

x1 = (x1_raw - x1_min) / (x1_max - x1_min)  # normalized x1
x2 = (x2_raw - x2_min) / (x2_max - x2_min)  # normalized x2

# True label
y_true = 85

# Initial model parameters
w1 = 0.6
w2 = 0.3
b = 10

# Learning rate
alpha = 0.001

# Step 1: Forward Pass - Predict the score
y_pred = w1 * x1 + w2 * x2 + b
print("Predicted Score ():", y_pred)

```
