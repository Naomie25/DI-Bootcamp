```python
import pandas as pd

df_2019 = pd.read_csv('/content/2019.csv')
print(df_2019.head())
```

```python
# Vérifier les valeurs manquantes
print(df_2019.isnull().sum())

# Supprimer ou remplir les valeurs manquantes
df_2019 = df_2019.dropna()

# Vérifier les types de données
print(df_2019.dtypes)

# Convertir au besoin

df_2019['Social support'] = df_2019['Social support'].astype(float)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(df_2019['Social support'], df_2019['Score'], color='green', alpha=0.7, edgecolors='black')
plt.title('Relation between social support and score of happiness (2019)', fontsize=14)
plt.xlabel('Social support')
plt.ylabel('Score of happiness')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# On suppose que df_2019 contient déjà la colonne 'Region'
region_group = df_2019.groupby('Country or region').mean(numeric_only=True).reset_index()

# Création du graphique combiné
fig, ax1 = plt.subplots(figsize=(12, 6))

# Barres pour le PIB
ax1.bar(region_group['Country or region'], region_group['GDP per capita'], color='skyblue', label='GDP per Capita')
ax1.set_ylabel('PIB par habitant', color='blue')
ax1.tick_params(axis='x', rotation=45)

# Deuxième axe pour l'espérance de vie
ax2 = ax1.twinx()
ax2.plot(region_group['Country or region'], region_group['Healthy life expectancy'], color='green', marker='o', label='Espérance de vie')
ax2.set_ylabel('Espérance de vie en bonne santé', color='green')

# Titre et layout
plt.title("Comparaison régionale : PIB et santé (2019)", fontsize=14)
plt.tight_layout()
plt.show()

```
