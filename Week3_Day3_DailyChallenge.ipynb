```python
import pandas as pd

df = pd.read_csv('/content/datascience_salaries.csv')
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df['salary_normalized'] = scaler.fit_transform(df[['salary']])
```

```python
from sklearn.decomposition import PCA

# Supprimer les colonnes non numériques et la cible (si besoin)
numeric_df = df.select_dtypes(include=['float64', 'int64']).drop(columns=['salary'])

# Appliquer PCA
pca = PCA(n_components=2)
reduced = pca.fit_transform(numeric_df)

# Ajouter les composantes au DataFrame
df['PC1'] = reduced[:, 0]
df['PC2'] = reduced[:, 1]
```

```python
# Afficher les moyennes et médianes par niveau d'expérience
average_salary = df.groupby('experience_level')['salary'].mean()
median_salary = df.groupby('experience_level')['salary'].median()

print("Salaire moyen par niveau d'expérience :\n", average_salary)
print("\nSalaire médian par niveau d'expérience :\n", median_salary)
```
