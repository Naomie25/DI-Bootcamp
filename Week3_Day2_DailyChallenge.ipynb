```python
from PIL import Image
import matplotlib.pyplot as plt

# Load an image using PIL
image_path = 'flowers/flowers/19_010.png'
original_image = Image.open(image_path)

# Display the original image using matplotlib
plt.imshow(original_image)
plt.axis('off')  # Optional: hides axis ticks
plt.show()
```

```python
from scipy.ndimage import rotate
def rotate_image_90_degrees(image):
    return rotate(image, angle=90, reshape=False)
```

```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
# Load and convert the image to an array
image = np.array(Image.open("flowers/flowers/19_010.png"))

# Rotate the image by 90 degrees
rotated_image = rotate_image_90_degrees(image)

# Display the rotated image
plt.imshow(rotated_image)
plt.axis('off')
plt.show()
```
